```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sensor_data.csv")
```

# Histogram of AccX

```python
plt.hist(df['AccX'], bins=20)
plt.xlabel('AccX')
```

# Scatterplot of GyroX vs GyroZ

```python
sns.scatterplot(data=df, x='GyroX', y='GyroZ')
```

# Violin plot of AccY by Class

```python
sns.violinplot(data=df, x='Target(Class)', y='AccY')
```

# Pairwise scatterplots

```python
sns.pairplot(df[['GyroX', 'GyroY', 'AccX', 'AccY']])
```

# 3D scatterplot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['GyroX'], df['GyroY'], df['AccZ'], c=df['Target(Class)'])
```

# Parallel coordinates plot

```python
cols = df.columns

plt.figure()
for i in range(len(df)):
  x = df.iloc[i, :]
  plt.plot(cols, x)
```
