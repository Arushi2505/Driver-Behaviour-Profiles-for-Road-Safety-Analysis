```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

#pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("test_motion_data.csv")
dataset.head()
```

```python
## print shape of the dataset with rows and columns
dataset.shape
```

```python
#1. Missing Values


## here we will check the percentage of nan values present in each feature

##  1. Step make the list of features which has missing values.

feature_with_na = [features for features in dataset.columns if dataset[features].isnull().any()]
## step print the feature name and the percentage of missing values
#print(feature_with_na)
for feature in feature_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

```python
# Measures of Central Tendency
print("AccX")
print("Mean =",  dataset.iloc[:,0].mean())
print("Median =",  dataset.iloc[:,0].mean())
print("Mode =",  dataset.iloc[:,0].mean()) 
print("--------------------------------------------------------")

print("AccY")
print("Mean =",  dataset.iloc[:,1].mean())
print("Median =",  dataset.iloc[:,1].mean())
print("Mode =",  dataset.iloc[:,1].mean())
print("--------------------------------------------------------")

print("AccZ")
print("Mean =",  dataset.iloc[:,2].mean())
print("Median =",  dataset.iloc[:,2].mean())
print("Mode =",  dataset.iloc[:,2].mean())
print("---------------------------------------------------------")

print("GyroX")
print("Mean =",  dataset.iloc[:,3].mean())
print("Median =",  dataset.iloc[:,3].mean())
print("Mode =",  dataset.iloc[:,3].mean())
print("---------------------------------------------------------")

print("GyroY")
print("Mean =",  dataset.iloc[:,4].mean())
print("Median =",  dataset.iloc[:,4].mean())
print("Mode =",  dataset.iloc[:,4].mean())
print("---------------------------------------------------------")

print("GyroZ")
print("Mean =",  dataset.iloc[:,5].mean())
print("Median =",  dataset.iloc[:,5].mean())
print("Mode =",  dataset.iloc[:,5].mean())
print("---------------------------------------------------------")

print("Timestamp")
print("Mean =",  dataset.iloc[:,7].mean())
print("Median =",  dataset.iloc[:,7].mean())
print("Mode =",  dataset.iloc[:,7].mean())
print("---------------------------------------------------------")
```

```python
#list of numerical variables

numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype !='O']   # O is the object

print('Number of numerical variables :', len(numerical_features))    

# visualise the numerical variables

dataset[numerical_features].head()     
```

```python
#dispersion
for col in dataset.columns:

# Range
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')
    print("Range: ", dataset[col].max() - dataset[col].min())

# Standard deviation
    print("Standard Deviation: ", dataset[col].std()) 

# Variance
    print("Variance: ", dataset[col].var())

# Coefficient of variation
    print("Coefficient of Variation: ", dataset[col].std() / dataset[col].mean()) 

# Interquartile range
    q1 = dataset[col].quantile(0.25)
    q3 = dataset[col].quantile(0.75)
    iqr = q3 - q1
    print("Interquartile range: ", iqr)
    print("------------------------------------------------------------------------------------")
```

```python
for col in dataset.columns:

  # Filter numeric data
  numeric_data = dataset[col][pd.to_numeric(dataset[col], errors='coerce').notnull()]

  # Check for numeric column
  if pd.to_numeric(dataset[col], errors='coerce').notnull().all():

    # Skewness and kurtosis
    print("Skewness:", dataset[col].skew())
    print("Kurtosis:", dataset[col].kurtosis()) 

    # Histogram plot 
    plt.hist(numeric_data, bins=20)
    plt.xlabel(col)
    plt.title("Histogram of "+col)
    plt.show()
    plt.close()

  else:
    print(col, "is non-numeric")

print("Shape analysis complete!")
```
